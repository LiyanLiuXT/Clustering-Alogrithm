```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial import Voronoi, voronoi_plot_2d,Delaunay

#points = np.array([[0, 0], [0, 1], [0, 2], [1, 0], [1, 1], [1, 2],[2, 0], [2, 1], [2, 2]])
points = np.array([[0, 0], [0, 1.1], [1, 0], [1, 1]])
vor = Voronoi(points)
voronoi_plot_2d(vor) # plot voronoi figure
tri = Delaunay(points)
plt.triplot(points[:,0], points[:,1], tri.simplices.copy())
plt.plot(points[:,0], points[:,1], 'o')
plt.show()


vertices = vor.vertices # get voronoi's vectices
print("vertices:", vertices)
```

```python
# plot Delaunay triangulation
tri = Delaunay(points)
plt.triplot(points[:,0], points[:,1], tri.simplices.copy())
plt.plot(points[:,0], points[:,1], 'o')
plt.show()
```

```python
triIndices = tri.simplices # get the point indices
print("get the point indices:\n",triIndices),print()

triCoordinates = points[tri.simplices] # get the point coordinates
print("get the point coordinates:\n", triCoordinates)
```
